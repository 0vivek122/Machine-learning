```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Price',axis =1), df['Price'], test_size=0.33, random_state=10)
```

# Feature Scaling so that we can reach the global minima faster

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train) #fit will caluate the mean ,standard deviation and then transfrom will apply fromula
```

X_new = (X - mean)/Std this is applies to every point of that column mean and Std(standard devistion of every single column)

```python
X_test =scaler.transform(X_test) #transform will convert using same mean and standard deviation used in train data
```

# Model

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
```

```python
l
```

```python
l.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
l.coef_
```

```python
## print intercept
l.intercept_
```

```python
##predict value of price for x_test
p=l.predict(X_test)
```

# Assumption for linear regression

```python
#the predict value and real value should follow linear pattern because there should be minimum or very less between them 
plt.ylabel('Predict value of Price')
plt.xlabel('Real value Price')
sns.scatterplot(y_test,p)
```

```python
#Residual difference between the actual and predicted value should be concentrated about zero
r=y_test-p #residual
sns.displot(r,kind="kde")
plt.xlabel('Error')
```

```python
#residual and prediction uniform distribution
sns.scatterplot(p,r)
plt.xlabel('Prediction')
plt.ylabel('Error')
```

# Perfromance Matrixs

```python
#calculating MSE,MAE,RMSE
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print('Mean squred error > ',mean_squared_error(y_test, p))
print('Mean Absolute error > ',mean_absolute_error(y_test, p))
print(' Root Mean squred error > ',n.sqrt(mean_squared_error(y_test, p)))
```

# R square and adjusted R square

```python
#R squared value
from sklearn.metrics import r2_score
r2=r2_score(y_test,p) #this mean 72 percent of data flow the linear regression equation
r2
```

```python
#Adjusted R sqaured value
1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1) #this give value according to effective features
```
